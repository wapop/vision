```python
yellow = cv2.imread("./img/yellow.jpg")
# cv2.imshow('yellow', yellow)
# cv2.waitKey(0)
```

```python
hsv = cv2.cvtColor(yellow, cv2.COLOR_BGR2HSV)
print(hsv.shape)
h = hsv[:,:,0]
print(h.shape)
```

```python
# 각 픽셀만큼 for loop를 돌면서 픽셀값을 처리  
for r in range(yellow.shape[0]):
    for c in range(yellow.shape[1]):
        # yellow[r,c,1] = [0,0,0]
        # yellow[r,c,2] = [0,0,0]

        if h[r, c] >= 27 and h[r, c] =< 32:
            yellow[r, c, :] = [255, 255, 255]
        else:
            yellow[r, c, :] = [0, 0, 0]
```

```python
imshow('yellow_binary', yellow)
```
