```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import subprocess as sp
%matplotlib inline

def imshow(tit, image) :
    plt.title(tit)    
    if len(image.shape) == 3 :
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    else :
        plt.imshow(image, cmap="gray")
    plt.show()
```

```python
green = cv2.imread('green.jpg')
imshow("green", green )   
```

```python
hsv = cv2.cvtColor(green, cv2.COLOR_BGR2HSV)
print(hsv.shape)
h = hsv[:,:,0]
s = hsv[:,:,1]
print(h.shape) 
print(s.shape)
```

```python
dest = green.copy()
for r in range(green.shape[0]) :
    for c in range(green.shape[1]) :
        if h[r,c] >= 50 and h[r,c] <= 100 and   s[r,c] >= 120 and s[r,c] <= 230 : 
            dest[r,c, :] =  green[r,c,:]  #[255,255,255]
        else :
            dest[r,c, :] = [0,0,0]         
# imshow("src", green)
# imshow("green", dest)

dest2 = np.hstack((green, dest ))
print(dest2.shape)
imshow("green", dest2)   
```
