```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

def imshow(tit, image) :
    plt.title(tit)    
    if len(image.shape) == 3 :
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    else :
        plt.imshow(image, cmap="gray")
    plt.show()
```

## cv2.imread(filename, flag)
- cv2.IMREAD_COLOR: 1
- cv2.IMREAD_GRAYSCALE :0 

```python
img = cv2.imread('lenna.bmp', 0)
print(img.shape)
print(img.dtype)
```

# point 연산
- y = alpha * Img + beta 
- clipping

```python
# 각 픽셀에 직접 접근하여 픽셀값 수정
# 속도 느림
dst = img.copy()
for r in range(img.shape[0]) :
    for c in range(img.shape[1]) :
            col = img[r,c] + 100
            if  col > 255  : col = 255
            dst[r,c] = col           
```

```python
imshow("img", np.hstack((img, dst) ))
```

```python
# numpy 연산
# 빠른 속도 but overflow 막지못함
dst = dst + 20
dst += 20
dst = img + 30
imshow("dst", dst)
```

## cv2.add()

```python
# 위에 비해 side effect도 없고 연산도 빠름
# cv2.add()
dst2 = cv2.add(img, 100)
imshow("dst", np.hstack((dst, dst2)))
```

```python
A = np.array([
    [1,5,],
    [10,20]
]).astype('uint8')
print(A)
print(A.dtype)
B = cv2.add(A, 240)
print(B)
```

```python
print(A[1,1] + 250 )    # 자동 형변환  uint8 -> int32
b = A + 250             # 자동 형변환이 되지만 b가 unit8 타입
c = A + 250.0           # 자동 형변환이 되고 c가 float 타입
d = A + 300             # 자동 형변환이 되고 d가 unit16 타입
print(b)
print(A.dtype)
print(b.dtype)
print(c.dtype)
print(d.dtype)
print(c)
```

```python
dd = (np.clip(A + 240.0, 0, 255)).astype('uint8')
print(dd)
print(dd.dtype)
```

## LUT cv2.LUT
- 비선형 변환
- 고속 변환

- 밝기 관점 아니라 그래프 조정 통한 rgb 3채널 색감 보정
- CF: https://blog.frame.io/2017/09/20/beginners-guide-to-curves/

```python
#반전
table = np.array([255-i for i in range(256)]) 
print(table)
```

```python
img1 = cv2.imread('lenna.bmp', 0)
out = cv2.LUT(img1, table) 
imshow("", out)
```

```python
dst = img1.copy()
for r in range(img1.shape[0]) :
    for c in range(img1.shape[1]) :
            dst[r,c] =   table[ img1[r,c] ] # 0 -> 255, 1 -> 254, ... 255 -> 0

imshow("",dst)
```

```python
# threshold LUT
table = np.array([ i > 127 for i in range(256)]) * 255
#print(table)
table = np.array([ 255 if i > 127 else 0 for i in range(256)])
#print(table)
out = cv2.LUT(img1, table) 
imshow("", out)
```

```python
p =100 / 255.0
gamma = 0.5
print(p, p ** gamma)     # gamm < 1  밝아짐

gamma = 1.0
print(p, p ** gamma)     #  gamma = 1.0 원본 그대로

gamma = 2.0
print(p, p ** gamma)     #  gamma > 1 어두워짐
```

```python
gamma = 0.5
table = np.array([((i / 255.0) ** gamma) * 255 
                  for i in np.arange(0, 256)]).astype("uint8")
#print(table)
plt.plot(table)
```

```python
files = ["gamma1.jpg", "gamma2.jpg", "gamma3.jpg", 'gamma4.jpg']
for f in files :
    img1 = cv2.imread(f)
    out = cv2.LUT(img1, table) 
    imshow("", np.hstack((img1,out)))
```

##  영상 합성 cv2.addWeighted

```python
img1 = cv2.imread('test1.jpg')
img2 = cv2.imread('test2.jpg')

alpha = 0.9
dst = cv2.addWeighted(img1, alpha, img2, 1-alpha, 0.0)

imshow("dst", dst)
```

```python
img1 = cv2.imread('test1.jpg')
img2 = cv2.imread('star.png')
 
img2 = cv2.resize(img2, dsize=(img1.shape[1],img1.shape[0]))  # 폭, 높이

bit_and = cv2.bitwise_and(img1, img2)

imshow("bit_and", bit_and)
```

```python
img1 = cv2.imread('test1.jpg')
img2 = cv2.imread('logo.png')

for r in range(img2.shape[0]) :
    for c in range(img2.shape[1]) :
            img1[r+0,c+0,:] = img2[r,c,:]
        
imshow("dst", img1)            
```

```python
img1 = cv2.imread('test1.jpg')
img2 = cv2.imread('logo.png')

for r in range(img2.shape[0]) :
    for c in range(img2.shape[1]) :
        if np.mean(img2[r,c,:]) > 20 :
            img1[r+0,c+0,:] = img2[r,c,:]
        
imshow("dst", img1)
```

```python
img1 = cv2.imread('test1.jpg')
img2 = cv2.imread('logo.png')

rows,cols,channels = img2.shape
roi = img1[0:rows, 0:cols ]
print(roi.shape)
print(img2.shape)
imshow("roi", roi)
imshow("img2", img2)
```

```python
img2gray = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
ret, mask = cv2.threshold(img2gray, 10, 255, cv2.THRESH_BINARY)
imshow('mask',mask)
```

```python
mask_inv = cv2.bitwise_not(mask)
imshow('mask_inv',mask_inv)
```

```python
img1_bg = cv2.bitwise_and(roi,roi,mask = mask_inv)
imshow('img1_bg',img1_bg)
```

```python
img2_fg = cv2.bitwise_and(img2,img2,mask = mask)
imshow('img2_fg',img2_fg)
```

```python
dst = cv2.add(img1_bg,img2_fg)
img1[0:rows, 0:cols ] = dst
imshow('res',img1)
```

```python
img1 = cv2.imread('gamma3.jpg', 0)
gamma = 0.5
table = np.array([((i / 255.0) ** gamma) * 255 
                  for i in np.arange(0, 256)]).astype("uint8")
out = cv2.LUT(img1, table)
imshow("src", img1)
imshow("gamma", out)

hist_1 = cv2.calcHist([img1],[0],None,[256],[0,256])
hist_2 = cv2.calcHist([out],[0],None,[256],[0,256])

plt.plot(hist_1, 'r')
plt.plot(hist_2, 'b')
```

## 히스토그램 스트레칭

```python
src = cv2.imread('lenna.bmp', 0 )
gmin = float(np.min(src))
gmax = float(np.max(src))
print(gmin, gmax)
```

```python
stretch =  ((src - gmin) * 255. / (gmax - gmin) ).astype(np.uint8)
sgmin = float(np.min(stretch))
sgmax = float(np.max(stretch))
print(sgmin, sgmax)
```

```python
hist_src = cv2.calcHist([src], [0], None, [256], [0,256])
hist_stretch = cv2.calcHist([stretch], [0], None, [256], [0,256])
plt.figure(figsize=(9, 6))
plt.plot(hist_src, 'r')
plt.plot(hist_stretch, 'b')
imshow('histogram stretching', np.hstack((src, stretch)))
```

```python
img = cv2.imread('cells.jpg.', 0)    
```

```python
from sklearn.cluster import KMeans
X = img.flatten().reshape(-1,1)
print(X.shape)
```

```python
model = KMeans(n_clusters=2, init="random", max_iter=20).fit(X)
p = model.cluster_centers_
print(p)
```

```python
th = (p[1,0] + p[0,0]) / 2.0
print(th)
```

```python
img = (img > th )*255
imshow("", img)
```
