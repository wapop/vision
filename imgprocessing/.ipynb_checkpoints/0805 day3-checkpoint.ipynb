```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import subprocess as sp
%matplotlib inline

def imshow(tit, image) :
    plt.title(tit)    
    if len(image.shape) == 3 :
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    else :
        plt.imshow(image, cmap="gray")
    plt.show()
```

# 1. 공간 필터링
### 인접 화소들의 값 참조하여 화소값 변경 처리
- 사칙연산아닌 연산 하는 비선형필터링
- ex 평균값 필터링

```python
src=cv2.imread('lenna.bmp',0)

#copy 보다 side effect 줄이려고, np.zeros(모두0인행렬만듦)
dst= np.zeros(src.shape, dtype=np.uint8)
#0이아니라 1로부터 시작하고, 맨끝 1앞부터 시작
for r in range(1, src.shape[0]-1):
    for c in range(1, src.shape[1]-1):
        sum= ( int(src[r,c-1]) + src[r,c] + src[r,c+1] ) /3  #나보다 하나왼, 나, 나보다하나오 더해서 overflow막기위해 평균값으로 나 결정
        dst[r,c] = sum
#src[r,c-1] : 타입모두 uint8 라서, 하나라도 int 타입으로 바꿔주기->확장
        
imshow("",np.hstack((src,dst)))

#블러 효과 더 크려면 주변 픽셀 더 늘리기
```

```python
# 3by3으로

src=cv2.imread('lenna.bmp',0)

dst= np.zeros(src.shape, dtype=np.uint8)
for r in range(1, src.shape[0]-1):
    for c in range(1, src.shape[1]-1):
        sum= 0
        for r1 in range(-1, 2):
            for c1 in range(-1,2):
                sum += src[r + r1, c +1]
        dst[r,c] = sum/9
        
imshow("",np.hstack((src,dst)))
sum= 0
# for r1 in range(-1, 2): 2까지 써야 바로앞1까지 기능
# for r in range(1, src.shape[0]-1): 1번째/마지막 픽셀은 내 위, 옆 픽셀이 없으니까!
```

```python
# 11by11으로
src=cv2.imread('lenna.bmp',0)

dst= np.zeros(src.shape, dtype=np.uint8)
for r in range(5, src.shape[0]-5):
    for c in range(5, src.shape[1]-5):
        sum= 0
        for r1 in range(-5, 6):
            for c1 in range(-5,6):
                sum += src[r + r1, c +1]
        dst[r,c] = sum/ (11*11)
        
imshow("",np.hstack((src,dst)))

#맨처음,맨마지막 고려해서 range 1-> 5으로 바꿔주고, sum 나눌때: 11*11로 나눠야함
```

```python
# 모션블러: r 지우면, 가로만 블러처리되게 됌! r은 가로고 c 세로
src=cv2.imread('lenna.bmp',0)

dst= np.zeros(src.shape, dtype=np.uint8)
for r in range(5, src.shape[0]-5):
    for c in range(5, src.shape[1]-5):
        sum= 0
        for c1 in range(-5,6):
            sum += src[r + r1, c +1]
        dst[r,c] = sum/ 11
        
imshow("",np.hstack((src,dst)))
```

```python
#빼기
src=cv2.imread('box.png',0)

dst= np.zeros(src.shape, dtype=np.uint8)
for r in range(1, src.shape[0]-1):
    for c in range(1, src.shape[1]-1):
        out= np.abs( int( src[r,c]) - src[r-1,c]) 
        dst[r,c]= out

dst= (dst>50)*255
dst= dst.astype('uint8')
imshow("",np.hstack((src,dst)))
#out= 0 아웃은 인티저타입, out=src[r,c]: out은 자기 자신, out= src[r,c] - src[r,c+1] 도 연산 안되니 int 바꾸기: -255~+255 사이 값
# src[r-1,c]
#주변픽셀 다 더하면 블러링, 옆만 하면 모션블러, out= np.abs( int( src[r,c]) - src[r-1,c]) 에서 r-1 하느냐/ c-1 하느냐 따라 수직선, 수평선 추출
```

```python
cv2.imshow("",np.hstack((src,dst)))
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
src=cv2.imread('letter.jpg',0)

dst= np.zeros(src.shape, dtype=np.uint8)
for r in range(1, src.shape[0]-1):
    for c in range(1, src.shape[1]-1):
        #out = np.abs(int(src[r,c]) - src[r,c+1]) # -255 ~ 255
        out= np.abs( int( src[r,c]) - src[r-1,c]) # -255 ~ 255    
        out1= np.abs( int( src[r,c]) - src[r,c-1]) # -255 ~ 255    
        dst[r,c]= (out+out1) / 2

dst= (dst>50)*255
dst= dst.astype('uint8')
imshow("",np.hstack((src,dst)))

cv2.imshow("",np.hstack((src,dst)))
cv2.waitKey()
cv2.destroyAllWindows()
```

## GaussianBlur 
- 가우시안 필터링 거리를 고려해주니까 평균필터링보다 많이 씀
- 표준편차를 크게 하면 주변 값 더 많이 참조: 표준편차로 블러링 정도 정해짐
- 표준편차 크면클수록 정규분포가 직선화 되니까 평균필터링(모든 데이터 가중치 동일)과 비슷해짐

```python
img= cv2.imread('lenna.bmp')
print(img.shape)
dst= cv2.blur(img, (7,7))
imshow("", dst)

dst1= cv2.GaussianBlur(img, (7,7), 1.0)
imshow("s=1.0", dst1)
dst1= cv2.GaussianBlur(img, (7,7), 20.0)
imshow("s=20.0", dst1)
```

## 샤프닝
- 3by3에서 가운데만 9이고, 나머지는 -1
- 보통 -1, -2 등 빼기가 있으면 같은걸 더해서 0으로 만들어줘야함
- embossing 에지 등 다 더해서 1이 나오는건 원래 영상을유지함 그러나 얘만 다름~
- <----> 더해서 0.8이면 전체영상이 어두워지는거고,
- 1.1이면 전체영상이 255넘어갈수있으니 제한 줘야함 (넘어가는 숫자에 대해서는 손실 일어날 수 있으니.)

```python
src= cv2.imread('rose.bmp', 0)
sharp = np.array([[-1, -1, -1], 
                  [-1, 9, -1],
                  [-1,-1,-1]], np.float32)

dst=cv2. filter2D(src, -1, sharp)
cv2.imshow('src', src)
cv2.imshow('dst', dst)

cv2.waitKey()
cv2.destroyAllWindows()
```

## noise
- 솔트앤페퍼형태의 노이즈는 medianBlur
- 바이너리 할때는 꼭 노이즈 없애주고 하기

```python
src= cv2.imread('lenna.bmp', cv2.IMREAD_GRAYSCALE)
cv2.imshow('src', src)

for stddev in [10, 20, 30]:
    noise = np.zeros(src.shape, np.int32)
    cv2.randn(noise, 0, stddev)
    
    dst= cv2.add(src, noise, dtype=cv2.CV_8UC1)
    
    desc= 'stddev= %d' % stddev
    cv2.putText(dst, desc, (10, 30), cv2. FONT_HERSHEY_SIMPLEX,
               1.0, 255, 1, cv2.LINE_AA)
    cv2.imshow('dst', dst)
    cv2.waitKey()

cv2.destroyAllWindows()
```

```python
import random

src= cv2.imread('lenna.bmp', cv2.IMREAD_GRAYSCALE)
cv2.imshow('src', src)

for i in range(0, int(src.size/10)):
    x= random.randint(0, src.shape[1]-1)
    y= random.randint(0, src.shape[0]-1)
    src[x, y] =(i%2)*255
    
    
dst1= cv2.GaussianBlur(src, (7,7), 1)
dst2= cv2.medianBlur(src,3)

ret, mask = cv2.threshold(dst2, 100, 255, cv2.THRESH_BINARY)
result = cv2.threshold(dst2, 100, 255, cv2.THRESH_BINARY)
(_, mask)=  = cv2.threshold(dst2, 100, 255, cv2.THRESH_BINARY)
#ret: result,, 의미없는,, 튜플이 생략된 형태

cv2.imshow('src', src)
cv2.imshow('dst1', dst1)
cv2.imshow('dst2', dst2)

cv2.waitKey()

cv2.destroyAllWindows()
```

## canny edge
- 컬러영상x 그레이에서 아웃라인 바이너리
- cv2.Canny(img_gray, 30, 180): 30보다 작으면 에지로 검출x, 180보다 크면 무조건 에지, 그 사이는 판단 알고리즘
- 불필요한 엣지 어떻게 날릴건지? --> 강한 에지랑 연결되있는지 아닌지로 판단 후 제거여부 결정
####
- canny 이미 잡음억제 위한 가우시안 적용되어있음, 추가로안해줘도 됨
- 파란색:수평, 빨간색: 수직, 노란/녹색:대각선 으로 방향 파악가능
- 두 개의 파라미터값(min, max) 특정 knowledge없으면 대략 1:2. 1:3사용

```python
img_gray= cv2.imread('result.jpg', cv2.IMREAD_GRAYSCALE)
img_canny= cv2.Canny(img_gray, 100, 500)
cv2.imshow("Canny Edge", img_canny)
cv2.waitKey()
cv2.destroyAllWindows()
```
